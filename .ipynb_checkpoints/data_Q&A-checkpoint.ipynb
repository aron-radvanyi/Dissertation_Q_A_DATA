```python
from docx import Document

import json

def convert_table_to_json(docx_file, table_index):
    doc = Document(docx_file)
    table = doc.tables[table_index]
    
    data = []
    
    # Loop through each row in the table and extract questions and answers
    for row in table.rows:
        question = row.cells[1].text.strip()
        answer = row.cells[2].text.strip()
        
        data.append({
            "prompt": question,
            "completion": answer
        })
    
    return json.dumps(data, indent=4)

word_file = "Word BAsed Q&A.docx"
table_index = 0  # Index of the table (0 for the first table, 1 for the second, and so on)

json_data = convert_table_to_json(word_file, table_index)
print(json_data)
```
